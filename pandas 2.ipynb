# Summary Statistics

###  Summarizing data, multiple summaries , Cumulative sum. 

Q1. Read the "Report" data

Q2. Find median of "Physics"

Q3. Find the eldest student

Q4. Use agg.() method to find (0.5) quantile on Maths column.

Q5. Use agg.() method on multiple columns(Maths & Physics) to find (0.5) quantile.

Q6. Use agg.() method on Maths column to find multiple Quantile (0.5 & 0.3).

Q7. Find cumulative sum of Maths. 
(Total 5 marks)

```python
# 1
## Start code here
import pandas as pd
report=pd.read_csv("report.csv",index_col=0)
print(report)
## End code here
```

```python
# 2
## Start code here
report["Physics"].median()
## End code here
```

```python
# 3
## Start code here
report["Date_of_birth"].min()
## End code here
```

```python
# 4
## Start code here
def pct50(column):
    return column.quantile(0.5)
report["Maths"].agg(pct50)
## End code here
```

```python
# 5
## Start code here
report[["Maths","Physics"]].agg(pct50)
## End code here
```

```python
# 6
## Start code here
def pct30(column):
    return column.quantile(0.3)
report["Maths"].agg([pct50,pct30])
## End code here
```

```python
# 7
## Start code here
report["Maths"].cumsum()
## End code here
```

## Pivot Tables

Q1. Pivot the Report Table by keeping index as "Place" & value as "Chemistry".

Q2. Using **numpy .agg** method find the median of the pivot table in question number 1.

Q3. Using **numpy .agg** method perform multiple statistics to find mean & median of the pivot table in question number 1.

```python
# 1
## Start code here
report.pivot_table(index="Place",values="Chemistry")
## End code here
```
